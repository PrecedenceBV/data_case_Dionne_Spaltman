```python
import pandas as pd 

# Define the file path
file_path = 'data-case/nps_scores.csv'

# Load the CSV file with ; as the separator
nps_scores = pd.read_csv(file_path, sep=';')

# Display the DataFrame
display(nps_scores)
```

```python
# Group by journey and calculate the mean NPS score
journey_nps = nps_scores.groupby('Journey_Type')['Recommendation_score'].mean().reset_index().sort_values(by='Recommendation_score')

print(journey_nps)
```

```python
print(nps_scores['Journey_Type'].unique())
```

```python
def calculate_nps(group):
    total = group.shape[0]
    promoters = group[group['Recommendation_score'] >= 9].shape[0]
    detractors = group[group['Recommendation_score'] <= 6].shape[0]
    nps = ((promoters - detractors) / total) * 100
    return nps

# Apply per journey
nps_per_journey = nps_scores.groupby('Journey_Type').apply(calculate_nps).reset_index(name='Recommendation_score')

# Find the lowest one
lowest_nps = nps_per_journey.sort_values(by='Recommendation_score')

print(lowest_nps)
```

```python

# Define the file path
file_path = 'data-case/call_data.csv'

# Load the CSV file with ; as the separator
call_data = pd.read_csv(file_path, sep=',')

# Display the DataFrame
display(call_data)
```

```python
import pandas as pd

# Assuming your dataframes are called call_data and nps_data
merged_df = pd.merge(call_data, nps_scores, on='Customer_ID', how='inner')

# Display the merged DataFrame
display(merged_df.head())
```

```python
# Group by journey and calculate average waiting time and average NPS score
journey_stats = merged_df.groupby('Journey_Type').agg({
    'Wait_Time_Minutes': 'mean',
    'Customer_Satisfaction_Score': 'mean',
    'Call_Duration_Minutes': 'mean',
}).reset_index()

# Sort by waiting time descending (longest wait first)
journey_stats = journey_stats.sort_values(by='Wait_Time_Minutes', ascending=False)

print(journey_stats)
```

```python
left_merged_df = pd.merge(call_data, nps_scores, on='Customer_ID', how='left')

display(left_merged_df)
```

```python
# Group by journey and calculate average waiting time and average NPS score
journey_stats = left_merged_df.groupby('Journey_Type').agg({
    'Wait_Time_Minutes': 'mean',
    'Customer_Satisfaction_Score': 'mean'
}).reset_index()

# Sort by waiting time descending (longest wait first)
journey_stats = journey_stats.sort_values(by='Wait_Time_Minutes', ascending=False)

print(journey_stats)
```

```python
# First, make sure call_resolved is lowercase for consistency
merged_df['Call_Resolved'] = merged_df['Call_Resolved'].str.lower()

# Group by journey and calculate stats
journey_stats = merged_df.groupby('Journey_Type').agg(
    total_calls=('Call_Resolved', 'count'),
    resolved_calls=('Call_Resolved', lambda x: (x == 'yes').sum()),
    unresolved_calls=('Call_Resolved', lambda x: (x == 'no').sum()),
    pct_resolved=('Call_Resolved', lambda x: (x == 'yes').mean() * 100),
    avg_waiting_time=('Wait_Time_Minutes', 'mean'),
    call_duration=('Call_Duration_Minutes', 'mean'),
    avg_nps_score=('Customer_Satisfaction_Score', 'mean')
).reset_index()

# Sort by pct_resolved descending (most resolved first)
journey_stats = journey_stats.sort_values(by='avg_nps_score', ascending=False)

display(journey_stats)
```

```python
import pandas as pd 

# Define the file path
file_path = 'data-case/order_data.csv'

# Load the CSV file with ; as the separator
order_data = pd.read_csv(file_path, sep=',')

# Display the DataFrame
display(order_data)
```

```python
import pandas as pd

# Assuming your dataframes are called call_data and nps_data
merged_df_order = pd.merge(order_data, nps_scores, on='Customer_ID', how='inner')

# Display the merged DataFrame
display(merged_df_order)
```

```python

# Group by journey and calculate stats
order_stats = merged_df_order.groupby('Journey_Type').agg(
    Order_Amount=('Order_Amount', 'mean'),
    avg_nps_score=('Recommendation_score', 'mean')
).reset_index()

# Sort by pct_resolved descending (most resolved first)
order_stats = order_stats.sort_values(by='avg_nps_score', ascending=False)

display(order_stats)
```

```python
import pandas as pd

# Calculate mean Order_Amount and avg NPS per Journey_Type
order_stats = merged_df_order.groupby('Journey_Type').agg(
    Order_Amount=('Order_Amount', 'mean'),
    avg_nps_score=('Recommendation_score', 'mean')
).reset_index()

# Calculate counts of each order_status per Journey_Type
order_status_counts = pd.crosstab(merged_df_order['Journey_Type'], merged_df_order['Order_Status'])

# Optional: Convert counts to percentages per Journey_Type
order_status_pct = order_status_counts.div(order_status_counts.sum(axis=1), axis=0) * 100

# Merge counts back to order_stats
order_stats = order_stats.merge(order_status_counts, left_on='Journey_Type', right_index=True)

# If you want, merge percentages too (with suffix _pct)
order_status_pct.columns = [f"{col}_pct" for col in order_status_pct.columns]
order_stats = order_stats.merge(order_status_pct, left_on='Journey_Type', right_index=True)

# Sort by avg_nps_score descending
order_stats = order_stats.sort_values(by='avg_nps_score', ascending=False)

display(order_stats)
```

```python
import pandas as pd

# Base stats: mean Order_Amount and avg NPS per Journey_Type
order_stats = merged_df_order.groupby('Journey_Type').agg(
    Order_Amount=('Order_Amount', 'mean'),
    avg_nps_score=('Recommendation_score', 'mean')
).reset_index()

# Count order_status per Journey_Type
order_status_counts = pd.crosstab(merged_df_order['Journey_Type'], merged_df_order['Order_Status'])
order_status_pct = order_status_counts.div(order_status_counts.sum(axis=1), axis=0) * 100
order_status_pct.columns = [f"order_status_{col}_pct" for col in order_status_pct.columns]

# Count Payment_method per Journey_Type
payment_method_counts = pd.crosstab(merged_df_order['Journey_Type'], merged_df_order['Payment_Method'])
payment_method_pct = payment_method_counts.div(payment_method_counts.sum(axis=1), axis=0) * 100
payment_method_pct.columns = [f"payment_method_{col}_pct" for col in payment_method_pct.columns]

# Count Product per Journey_Type
product_counts = pd.crosstab(merged_df_order['Journey_Type'], merged_df_order['Product'])
product_pct = product_counts.div(product_counts.sum(axis=1), axis=0) * 100
product_pct.columns = [f"product_{col}_pct" for col in product_pct.columns]

# Merge all counts and percentages back to order_stats
order_stats = order_stats.merge(order_status_counts, left_on='Journey_Type', right_index=True)
order_stats = order_stats.merge(order_status_pct, left_on='Journey_Type', right_index=True)

order_stats = order_stats.merge(payment_method_counts, left_on='Journey_Type', right_index=True)
order_stats = order_stats.merge(payment_method_pct, left_on='Journey_Type', right_index=True)

order_stats = order_stats.merge(product_counts, left_on='Journey_Type', right_index=True)
order_stats = order_stats.merge(product_pct, left_on='Journey_Type', right_index=True)

# Sort by avg_nps_score descending
order_stats = order_stats.sort_values(by='avg_nps_score', ascending=False)

display(order_stats)
```
